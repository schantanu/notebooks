### Wordle Letter and Position frequency

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt

# Create empty dataframe for of each letter in each position
letter_frequency_df = pd.DataFrame(0, index=range(1, 6), columns=list('abcdefghijklmnopqrstuvwxyz'))

# Count letter frequency by letter and position
for word in words:
    for position, letter in enumerate(word):
        letter_frequency_df.at[position + 1, letter] += 1

# Calculate totals for each letter
letter_frequency_df.loc["Total"] = letter_frequency_df.sum()
letter_frequency = letter_frequency_df.T

# Create a grid for the heatmaps
grid = plt.GridSpec(1, 2, width_ratios=[6, 2], wspace=0.1)

# Set plot size
plt.figure(figsize=(8, 8))

# Plot the data heatmap
plt.subplot(grid[0, 0])
sns.heatmap(letter_frequency.iloc[:, :-1], annot=True, fmt='g', cmap="crest", cbar=False)
plt.yticks(rotation=0)
plt.title('Frequency by \nLetter and Position')

# Plot the totals heatmap
plt.subplot(grid[0, 1])
sns.heatmap(letter_frequency[['Total']], annot=True, fmt='g', cmap="Oranges", cbar=False).yaxis.tick_right()
plt.yticks(rotation=0)
plt.title('Frequency by \nLetter')

plt.show()
```

### Top 10 Wordle starter words

```python
# Rank words based on frequency of letters and their positions
def rank_word(word):
    score = 0
    for position, letter in enumerate(word):
        score += letter_frequency_df.at[position + 1, letter]
    return score

# Create word score dictionary
word_scores = {word: rank_word(word) for word in words}

# Sort words by score
sorted_words = sorted(word_scores.items(), key=lambda x: x[1], reverse=True)

# Show top 10 words by score
print("Top 10 starting words based on letter frequency and positions:")
for word, score in sorted_words[:10]:
    print(f"{word}: {score}")
```
